# task for all movies title from the main page subslikescript

```python
import requests 
from bs4 import BeautifulSoup

website="https://subslikescript.com/movies"
result=requests.get(website)
content=result.text
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
list=soup.find("ul",class_="scripts-list")
print(list.text)
```

```python
x=list.find_all("a")
website="https://subslikescript.com"
name="" 
for temp in x:
    name=temp.text
    temp=str(temp)
    s=temp.find('"')
    site=temp[(s+1):]
    site=site[:site.find('"')]
    temp_site=website+site
    result=requests.get(temp_site)
    content=result.text
    soup=BeautifulSoup(content,'html.parser')
    box=soup.find('article')
    script=box.find('div',class_='full-script').text
    f=open(name.strip(),'w',encoding='utf-8')
    f.write(script)
```

# rottentomatoes task

```python
import requests 
import pandas as pd
from bs4 import BeautifulSoup

website="https://www.rottentomatoes.com/browse/movies_in_theaters/sort:popular"
result=requests.get(website)
content=result.text
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
movie=[]
rating=[]
box=soup.find_all("div",class_="flex-container")

# print(movies)
for i in box:
    movies=soup.find("span",class_="p--small")
    movies=i.text.strip()
    movie.append(movies)
    percentage=soup.find("rt-text")
    percentage=i.text.strip()
    rating.append(percentage)
print("MOvies Name:",movie)
print("Movies Rating:",rating)
print(len(movie))
print(len(rating))
```

```python
data=pd.DataFrame()
data["Movies Name"]=movie
data["rating"]=0
data
```
